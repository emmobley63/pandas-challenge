```python
#Import Dependencies

import pandas as pd

#Create access to csv

datafile = "Resources/purchase_data.csv"

df = pd.read_csv(datafile)
df.head()
```

```python
# Count the number of unique SN's to find the Total Number of Players

totplayers = df["SN"].nunique() 
totplayers_df =pd.DataFrame({"Total Players": totplayers}, index = [0])

totplayers_df
```

```python
# Purchasing Analysis
# Count the number of unique purchase items

uitems = df["Item Name"].nunique()
uitems

#Average purchase price, Total number of purchases, Sum of price for total revenue

avgprice = df["Price"].mean()
totpurchases = len(df["Purchase ID"])
totrev = df["Price"].sum()

tot_df = pd.DataFrame({"Number of Unique Items": uitems,
                       "Average Price": avgprice,
                       "Number of Purchases": totpurchases,
                       "Total Revenue": totrev}, index = [0])
tot_df
# If I want to put into table I think that I have to enter the data into a DF
```

```python
# Male, Female and non-disclosed counts by groupby

maledf = df.loc[df["Gender"] == "Male"]

malecount = len(maledf.groupby("SN"))

femaledf = df.loc[df["Gender"] == "Female"]

femalecount = len(femaledf.groupby("SN"))

nddf = df.loc[df["Gender"] == "Other / Non-Disclosed"]

ndcount = len(nddf.groupby("SN"))

# Percent by gender

mper = malecount / totplayers * 100
fper = femalecount / totplayers * 100
ndper = ndcount / totplayers * 100

gender_df = pd.DataFrame({"Total Count": [malecount, femalecount, ndcount],
                          "Percentage of Players":[mper, fper, ndper]})
gender_df
```

```python
# gendergroup = df.groupby("Gender")

# print(gendergroup["Purchase ID"].count(), \
# gendergroup["Price"].sum())

# Purchase count of M/F/ND of purchases

mpc = len(maledf)
fpc = len(femaledf)
ndpc = len(nddf)

print(mpc, fpc, ndpc)
```

```python
# Total purchase value by gender

mtpv = maledf["Price"].sum()
ftpv = femaledf["Price"].sum()
ndtpv = nddf["Price"].sum()

print(mtpv, ftpv, ndtpv)
```

```python
# Average purchase price by gender

mapp = mtpv / mpc
fapp = ftpv / fpc
ndapp = ndtpv / ndpc

print(mapp, fapp, ndapp)
```

```python
#Average Total Purchase per Person

matpp = mtpv / malecount
fatpp = ftpv / femalecount
ndatpp = ndtpv / ndcount

print(matpp, fatpp, ndatpp)
```

```python
genpur_df = pd.DataFrame({"Purchase Count": [mpc, fpc, ndpc],
                         "Average Purchase Price": [mapp, fapp, ndapp],
                         "Total Purchase Value": [mtpv, ftpv, ndtpv],
                         "Avg Total Purchase per Person": [matpp, fatpp, ndatpp]})
genpur_df
```

```python
# Age Demographics

# Establish bins and labels for ages
age_splits = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 130]

age_group = pd.cut(df["Age"], bins, labels=age_splits)

age_group.head()
```

```python
# Add Age Group column to original dataframe

df["Age Group"] = age_group
df.head()
```

```python
# Total Count and Percentages by age group and percentage by age group

agegroupby = df.groupby("Age Group")


count_per_age = agegroupby["SN"].nunique()

age_per = [((x/totplayers)*100) for x in count_per_age]

count_per_age, age_per
```

```python
# Create new DataFrame to Display the Table

age_df = pd.DataFrame({"Total Count": count_per_age,
                      "Percentage of Players": age_per})
age_df
```

```python
# Groupby age groups
agegroupby = df.groupby("Age Group")

# Establishing lists to be placed in chart

age_pc = agegroupby["Purchase ID"].count() 
age_app = agegroupby["Price"].mean() 
age_tpv = agegroupby["Price"].sum() 


age_aptpp = age_tpv / count_per_age

purchase_df = pd.DataFrame({"Purchase Count": age_pc,
                           "Average Purchase Price": age_app,
                           "Total Purchase Value": age_tpv,
                           "Avg Total Purchase per Person": age_aptpp})
purchase_df
```

```python
#Group by name (SN)

sngroupby = df.groupby("SN")

# Establish variable lists for the DF

sn_pc = sngroupby["Purchase ID"].count()
sn_app = sngroupby["Price"].mean()
sn_tpv = sngroupby["Price"].sum()

# Create DF

sn_df = pd.DataFrame({"Purchase Count": sn_pc,
                     "Average Purchase Price": sn_app,
                     "Total Purchase Value": sn_tpv})

# Sort by descending Total Purchase Value

sn_df.sort_values("Total Purchase Value", ascending=False).head(5)
```

```python
#Most Popular Items

# Retrieve the Item ID, Item Name, and Item Price columns
item_df = df[["Item ID","Item Name","Price"]]

item_df.head()
```

```python
# Groupby ID and Item name
itemgroupid = item_df.groupby("Item ID")

itemgroupid.head()

#Purchase Count, Average Item Price, and Total Purchase Value

item_count = itemgroupid["Item Name"].count()
item_tpv = itemgroupid["Price"].sum()
item_ap = itemgroupid["Price"].mean()

item_df = pd.DataFrame({"Purchase Count": item_count,
                       "Item Price": item_ap,
                       "Total Purchase Value": item_tpv})
item_df.sort_values("Purchase Count", ascending=False)
```
